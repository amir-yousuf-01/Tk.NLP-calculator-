```python
import tkinter as tk
import  re
from tkinter import messagebox
```

```python
def prase_math(text):
    text = text.lower()

    text = text.replace("plus","+").replace("add","+")
    text = text.replace("minus","-").replace("subract","-")
    text = text.replace("times","*").replace("multiplied","*")
    text = text.replace("divide","/").replace("divided by","/")


    pattern = r"(\d*\.?\d*)\s*([\+\-\*/])\s*(\d*\.?\d*)"
    match = re.search(pattern, text)
    if match:
        num1, op, num2 = match.groups()
        try:
            num1 = float(num1) if num1 else 0
            num2 = float(num2) if num2 else 0
            if op == '+':
                return num1 + num2
            elif op == '-':
                return num1 - num2
            elif op == '*':
                return num1 * num2
            elif op == '/':
                if num2 == 0:
                    return "Division by zero"
                return num1 / num2
        except ValueError:
            return "Invalid number format"
    else:
        return"Could not Prase input"

    
 


def calculate():
    user_input = entry.get().strip()
    if not user_input:
        messagebox.showerror("Eror", "Input field is empty")
        return
    result = prase_math(user_input)
    if isinstance(result,str):
        messagebox.showerror("Error",result)
    else:
        messagebox.showinfo("Result",f"Result:{result:.2f}")
        result_var.set(f"{result:.2f}")


def clear():
    entry.delete(0,tk.END)
    result_var.set("0")


root = tk.Tk()
root.title("NLP Calculator")
root.geometry("400x300")


result_var = tk.StringVar()
result_var.set("0")
display = tk.Label(root, textvariable=result_var, font=("Arial", 24), anchor="e", bg="white", width=20)
display.pack(pady=20,padx=10, fill = "x")

label = tk.Label(root,text ="Enter calculation (e.g., 'add 5 and 7' or '5+3'):")
label.pack()


entry = tk.Entry(root,width=50,font=("Arial",14))
entry.pack(pady=10)
entry.bind("<Return>", lambda event: calculate())


calc_button = tk.Button(root,text ="Calculate", command = calculate, font =("Arial", 14))
calc_button.pack(pady=5)

clear_button = tk.Button(root, text="Clear",command=clear, font= ("Arial", 14) )
clear_button.pack(pady=5)

root.mainloop()
```
